This demo shows how to read the AR1.csv data set into Python. AR1.csv is on my google drive. ACF (autocorrelation plot) is then plotted to check if the process is independent or not.

```python
import math
import statistics
import numpy as np
import scipy.stats
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# mount google drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
#the 2nd read option assume that the data file is rignt under your default google drive directory
# the sep-',' argument in the read statement should be used in some cases if ; does not work
data = pd.read_csv('drive/My Drive/Colab Notebooks/IMSE_Data_Science/Data/AR1.csv', sep=',', na_values=".")
data.head() 
```

```python
# is this time series stationary?
plt.plot(data)
```

```python
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(data, lags=20)
```

# if the time series data is independent, there should not be any spikes beyond the shadded area after lag 0.

```python
# ACF plot for an independently distribution time series
sample_size = 100
data2 = np.random.normal(size=sample_size)
plt.plot(data2)
plot_acf(data2, lags=20)
```

The above ACF plot show that all spikes are small after lag 0 (the largest pike at lag0). Therefore, the data is independent.
