```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# a sample dataset about cars
cars = sns.load_dataset('mpg')
```

```python
type(cars)
```

```python
cars.head()
```

```python
cars. shape
```

```python
#delete the row with na (null) data
cars.dropna(inplace=True)

cars.shape
```

```python
# matplotlib demo using histogram
# ref https://matplotlib.org/stable/gallery/statistics/hist.html

plt.hist(x=cars.mpg, bins=20)
plt.show()
```

```python
# use seaborn to improve the matplot plots
sns.set() 
plt.hist(x=cars.mpg, bins=20)
plt.show()
```

```python
# plot bar charts in different panels
# the plots are in one row and three columns using the subplots
# each plot is a hist() plot
dist1 =cars.mpg
dist2 =cars.cylinders
dist3 =cars.acceleration
n_bins=20

fig, axs = plt.subplots(1, 3, sharey=True, tight_layout=True)

# We can set the number of bins with the *bins* keyword argument.
axs[0].hist(dist1, bins=n_bins)
axs[1].hist(dist2, bins=n_bins)
axs[2].hist(dist3, bins=n_bins)
```

```python
sns.relplot(x='model_year', y='mpg', col='origin', hue='cylinders', 
            data=cars);
```

```python
# x is used for the sections
# y is the outcome for comparison
y=cars['mpg']
x=cars['cylinders']
sns.boxplot(x, y)
```

```python
# a short cut of the previous two box
sns.boxplot(x=cars.cylinders, y=cars.mpg)
```
