```
# Importing Necessary Libraries
import numpy as np
from skimage import data
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
```

```
# Setting the plot size to 15,15
plt.figure(figsize=(15, 15))
 
# Sample Image of scikit-image package
coffee = data.coffee()
plt.subplot(1, 2, 1)
 
# Displaying the sample image
plt.imshow(coffee)
# Converting RGB image to Monochrome
gray_coffee = rgb2gray(coffee)
plt.subplot(1, 2, 2)
 
# Displaying the sample image - Monochrome
# Format
plt.imshow(gray_coffee, cmap="gray")
```

Explanation: By using rgb2gray() function, the 3-channel RGB image of shape (400, 600, 3) is converted to a single-channel monochromatic image of shape (400, 300). 

# Convert the above grayscale image so that the image looks like a negative, i.e. white becomes black and black becomes write

```
gray_coffee.shape
```

```
for value in gray_coffee:
  print(value)
```

```
nr=int(len(gray_coffee.mean(axis=1)))
nc=int(len(gray_coffee.mean(axis=0)))
print("number of rows=", nr)
print("number of columns=", nc)
```

```
# prepare a new image
# this all zeros in gray image will show all black since the values are zeros
negative_coffee=np.zeros_like(gray_coffee)
# since the gray scale has been standardized in [0, 1], 0 - black; 1 - white
# try: negative_coffee[i][j]= -gray_coffee[i][j]
for i in range(0, nr):
    for j in range(0, nc):
        negative_coffee[i][j]=1-gray_coffee[i][j]
    
# plt.imshow(negative_coffee, cmap="gray")

# Displaying the images in two panels
# Setting the plot size to 15,15
plt.figure(figsize=(15, 15))
plt.subplot(1, 2, 1)
plt.imshow(gray_coffee,cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(negative_coffee, cmap="gray")
```

# What if you add the images?

```
# prep for a new image
add_coffee=np.zeros_like(gray_coffee)
# since the gray scale has been standardized in [0, 1], 0 - black; 1 - white
for i in range(0, nr):
    for j in range(0, nc):
        add_coffee[i][j]= negative_coffee[i][j] + gray_coffee[i][j]
        
plt.imshow(add_coffee, cmap="gray")
```

Since the origial and negative pixel values add up to zeros, so it a black everywhere.
