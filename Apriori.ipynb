# The code domonstrates the use of mixtend library for the apriori algorithm

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
# there are six items in 16 transactions
# item 1 is index as (0), item 2 as (1) and so on
dataset = [['1', '2', '3'],
           ['2', '3', '4'],
           ['3','4','5'],
           ['4', '5', '6'],
           ['1','3','5'],
		   ['2', '4', '6'],
           ['1','3','4'],
		   ['2', '4', '5'],
           ['3','5','6'],
		   ['1', '2', '4'],
           ['2','3','5'],
           ['3','4','6'],
           ['1','2'],
           ['1','2'],
           ['1','2'],
           ['1','2']]
```

```python
dataset2=[['A', 'B', 'C', 'D'],
          ['A', 'B', 'C'],
          ['A', 'B'],
          ['B','C'],
          ['A'],
          ['B']]
```

```python

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
# Executing the apriori algorithm

apriori(df, min_support=0.36)
```

```python

te = TransactionEncoder()
te_ary = te.fit(dataset2).transform(dataset2)
df2 = pd.DataFrame(te_ary, columns=te.columns_)
#df2
#from mlxtend.frequent_patterns import apriori

apriori(df2, min_support=0.50)
```
