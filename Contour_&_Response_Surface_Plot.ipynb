# Second Order Model Contour Plot & Response Surface Plot
by SIC K-State 10/25/2023

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated data
x1 = np.linspace(0, 10, 100)
x2 = np.linspace(0, 10, 100)
y = 3 * x1 + 2 * x2 + (x1-5)*(x1-5) + (x2-4)*(x2-4) + 5 + np.random.normal(0, 2, 100)
```

Your Turn: see if you can fit a 2nd order regression model on y

# Model Estimation & Generating Grid data points

```python
# Create a 2D grid of x1 and x2 values
X1, X2 = np.meshgrid(x1, x2)

# Calculate the response variable (y_pred) using your model or equation
# the following equation assumes your have a perfect estimation
# use the fitted betas into the following equation
y_pred = 3 * X1 + 2 * X2 + 5 + (X1-5)*(X1-5) + (X2-4)*(X2-4)
```

#2D Contour Plot

```python
# Create an contour plot
plt.figure(figsize=(8, 6))
plt.contourf(x1, x2, y_pred, cmap='viridis')
plt.colorbar()
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Contour Plot')
plt.show()
```

# Response Surface Plot (3D Plot)

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X1, X2, y_pred, 50, cmap='binary')
ax.set_xlabel('x1)')
ax.set_ylabel('x2')
ax.set_zlabel('y_pred');
```
