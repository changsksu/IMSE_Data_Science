```python
import pandas as pd
import seaborn as sns
```

```python
# load a sample dataset called cars
cars = sns.load_dataset('mpg')
```

```python
# generate the descriptive statistics of this data set
cars.describe()
```

```python
# cars is a dataframe
cars.head()
```

```python
# get the first 10 rows
cars[0:9]
```

```python
# get the first 2 columns
# don't ignore the 2nd set of brackets
cars[['mpg','cylinders']]
```

```python
# get the first two columns and 10 rows
cars.iloc[0:10, 0:2]
```

```python
# insert a new column
cars['kpl']=cars['mpg'] * 1.609 / 3.785
cars.head()
```

```python
# sort the data by model year in decending order
# note that we did not save it back to replace the df cars

cars.sort_values(by='model_year', ascending=False)
```

```python
# simple XY plots
cars.plot('model_year', 'mpg')
```
