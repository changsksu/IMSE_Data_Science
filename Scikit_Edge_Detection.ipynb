# Scikit Example of edge detection

```
from skimage import data, io, filters
#coin image
image = data.coins()
io.imshow(image)
```

```
# ... or any other NumPy array!
edges = filters.sobel(image)
io.imshow(edges)
io.show()
```

```
edges.shape
```

```
edges
```

```
first_coin=edges[30:75, 20:75]
io.imshow(first_coin)
io.show()
```
