```python
import numpy as np
import matplotlib.pyplot as plt

def cltexp(n, lam):
    y = np.zeros(500)  # create array of zeros length 500
    for i in range(500):
        y[i] = np.mean(np.random.exponential(scale=1/lam, size=n))

    plt.hist(y, bins=20, edgecolor="black")
    plt.title("Sampling Distribution of the Mean (Exponential)")
    plt.xlabel("Sample mean")
    plt.ylabel("Frequency")
    plt.show()
```

```python
cltexp(1,0.5)
```

```python
cltexp(10,0.5)
```
